```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# Upload and read data
from google.colab import drive
drive.mount("/content/drive")
path = "/content/drive/MyDrive/Loan_data_vis_RBL.xlsx"
data = pd.read_excel(path, engine="openpyxl")

# Filter the data to include only rows where 'Loan Approval' is 1
approved_loans = data[data['LoanApproved'] == 1]

# Group the data by 'Education Level' and 'Loan Purpose' and count the number of approved loans
loan_approval_counts = approved_loans.groupby(['EmploymentStatus', 'EducationLevel', 'MaritalStatus', 'HomeOwnershipStatus', 'LoanPurpose']).size().unstack(fill_value=0)

# Plot the bar chart
loan_approval_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Adding title and labels
plt.title('Loan Approval Counts')
plt.xlabel('factors')
plt.ylabel('Count of Loan Approvals (1)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='factors')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to avoid clipping of labels
plt.show()
```
